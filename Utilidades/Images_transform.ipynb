```python
#@title Funciones generadas:
##Función cuyo proposito es redimensionar las imagenes al alto y ancho deseado:
def images_transform(Origin, Destiny, width, height, type = True):
  for m in os.listdir(Origin):
    if m != '.ipynb_checkpoints':
      name = Origin + '/' + m
      img = cv2.imread(name)
      print(name)
      img_new = cv2.resize(img,(width,height))
      if type == True:
        plt.figure(figsize=(15,10))
        plt.subplot(1,2,1)
        plt.title('Imagen Original')
        plt.axis('off')
        plt.imshow(img[:,:,(2,1,0)]) ##Se invierten los canales.
        plt.axis()
        plt.subplot(1,2,2)
        plt.title('Imagen Redimensionada')
        plt.imshow(img_new[:,:,(2,1,0)]) ##Se invierten los canales.
        plt.axis('off')
        plt.show()
        break
      elif type == False:
        name_des = Destiny + '/' + m
        #print(name_des)
        cv2.imwrite(name_des,img_new)
```

```python
#@title Se visualiza una imagen redimensionada
directorio='/content/drive/MyDrive/YOLO_CUSTOM_DATASE/images_ori'
destino ='/content/drive/MyDrive/YOLO_CUSTOM_DATASE/images'
height = 416
width = 416
images_transform(directorio, destino,width,height)
```
